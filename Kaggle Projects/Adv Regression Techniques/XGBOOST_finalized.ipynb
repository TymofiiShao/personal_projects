```python
import pandas as pd
import numpy as np
```

```python
df_valid = pd.read_csv('DF_TRAINED_PREPROCESSED_RANDOM_FOREST_2.csv')
df_valid.head()
```

## Model

```python
X = df_valid.drop(['Id', 'SalePrice'], axis = 1)
y = df_valid.SalePrice
```

```python
from sklearn.model_selection import train_test_split

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.2)
```

```python
import xgboost as xgb
```

```python
model = xgb.XGBRegressor()
model = model.fit(X_train, y_train)
y_pred = model.predict(X_valid)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

rmse = np.sqrt(mean_squared_error(y_valid, y_pred))
mae = mean_absolute_error(y_valid, y_pred)
r2 = r2_score(y_valid, y_pred)

y_pred = model.predict(X_valid)
print('RMSE:', rmse)
print('MAE:', mae) 
print('R2:', r2) 
```

```python
y_pred = model.predict(X_test) 
```

## Grid Search

```python
from xgboost import cv
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],  # кількість дерев у ансамблі
    'colsample_bytree': [0.3, 0.5, 0.7],  # доля ознак при побудові дерева
    'learning_rate': [0.01, 0.1, 0.2],  # швидкість навчання
    'max_depth': [5, 10, 15],  # максимальна глибина дерев
    'alpha': [5, 10, 15]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_search = GridSearchCV(estimator = model, param_grid=param_grid, cv=10, 
                           scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)
grid_search.fit(X_train, y_train)
```

```python
best_model = grid_search.best_estimator_
```

```python
print("Найкращі параметри:", grid_search.best_params_)
```

```python
model_2 = xgb.XGBRegressor(alpha=5, colsample_bytree=0.3, learning_rate=0.2, max_depth=5, n_estimators=100)
model_2 = model_2.fit(X_train, y_train)
y_pred_2 = model_2.predict(X_valid)
```

```python
rmse_2 = np.sqrt(mean_squared_error(y_valid, y_pred_2))
mae_2 = mean_absolute_error(y_valid, y_pred_2)
r2_2 = r2_score(y_valid, y_pred_2)

y_pred_2 = model_2.predict(X_valid)
print('RMSE:', rmse_2)
print('MAE:', mae_2) 
print('R2:', r2_2) 
```

## Feature Importance

```python
import matplotlib.pyplot as plt
import seaborn as sns

feature_importance = model_2.feature_importances_
feature_names = X.columns
feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': feature_importance})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)
top_features = feature_importance_df.head(10)
```

```python
total_importance = top_features['importance'].sum()
top_features['importance_percentage'] = (top_features['importance'] / total_importance) * 100
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='feature', y='importance_percentage', data=top_features, palette='viridis')
plt.title('Top 10 Feature Importance (%)')
plt.xlabel('Feature')
plt.ylabel('Importance (%)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
